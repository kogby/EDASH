```python
import pandas as pd
import numpy as np
import math
import statsmodels.api as sm
from matplotlib import pyplot as plt

# Experimental packages
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
```

```python
def imputation_OLS(df, method, target, mask=[]):
    df = df.drop(mask, axis=1)
    df_o =df
    df = df.dropna()
    X = df.drop(target,axis=1)
    Y = df[target].values
    X = sm.add_constant(X)
    model = sm.OLS(Y,X).fit()
    print(f"The original R^2 = {model.rsquared:.4f}")
    print(f"The original adj-R^2 = {model.rsquared_adj:.4f}")
    
    df_imp = EM_algorithm(df_o)
    X_imp = df_imp.drop(target,axis=1)
    X_imp = sm.add_constant(X_imp)
    Y_imp = df_imp[target].values
    model_imp = sm.OLS(Y_imp,X_imp).fit()
    print(f"The original R^2 = {model_imp.rsquared:.4f}")
    print(f"The original adj-R^2 = {model_imp.rsquared_adj:.4f}")
```

```python
def EM_algorithm(df):
    imputer = IterativeImputer(random_state=1105)
    imputer.fit(df)
    df_imp = imputer.transform(df)
    df_imp = pd.DataFrame(df_imp,columns=df.columns)
    display(df_imp)
    return df_imp
```
